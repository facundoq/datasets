```python
basepath='/home/facundo/datasets/rwth-boston-104'
output_path='/home/facundo/datasets/rwth-boston-104-split'
import os
if not os.path.exists(output_path):
    os.mkdir(output_path)


import metadata

all_indices=range(1,201)
test_indices=[2,7,12,21,25,28,30,36,40,43,50,54,57,67,71,74,77,84,89,90,92,
              100,105,107,108,113,119,122,139,142,158,167,171,174,181,184,189,193,199,201]
train_indices= list(set(all_indices) - set(test_indices))

input_video_positions_filepath=os.path.join(basepath,'handpositions/full.xml')
train_output_video_positions_filepath=os.path.join(basepath,'handpositions/train.xml')
test_output_video_positions_filepath=os.path.join(basepath,'handpositions/test.xml')
```

```python



def generate_subset(input_video_positions_filepath,output_video_positions_filepath,indices):
    
    from xml.dom import minidom
    xmldoc = minidom.parse(input_video_positions_filepath)
    positions_element=xmldoc.firstChild
    video_elements = positions_element.getElementsByTagName('video')
    video_frame_positions=[]
    for video_element in video_elements:
        video_id = int(video_element.attributes['name'].value)
        if not video_id in indices:
            print("removing video %d" % video_id)
            positions_element.removeChild(video_element)
    xml_string=xmldoc.toxml()
    with open(output_video_positions_filepath, "w") as text_file:
        print(xml_string, file=text_file)
    
print("Generating file %s with info about %d videos "% (train_output_video_positions_filepath,len(train_indices)))    

generate_subset(input_video_positions_filepath,train_output_video_positions_filepath,train_indices)
print("Generating file %s with info about %d videos "% (test_output_video_positions_filepath,len(test_indices)))    
generate_subset(input_video_positions_filepath,test_output_video_positions_filepath,test_indices)


```
