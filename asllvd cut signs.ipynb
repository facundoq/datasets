```python
import os
#import numpy as np
#import skimage.io
#import skimage
#from scipy import ndimage

import subprocess

import skvideo.io
#import skvideo.datasets

import csv

```

```python
# csv file with the annotations, converted from the excel file at http://www.bu.edu/av/asllrp/dai-asllvd.html
annotations_path='/media/data/datasets/asllvd/annotations.csv' 
# folder with all the videos from the camera 1 of the dataset, uncut, with filenames in the format:
#  SESSION_SCENE_camera1.mov
videos_path='/media/data/datasets/asllvd/videos'
# folder where 
output_path='/media/data/datasets/asllvd/segmented_videos'

# minimum duration for cut signs. If the annotations imply that a sign lasts for less that this minimum
# duration, the script will extend the duration uniformly from the start and end of the annotations.
min_duration=0.3 

if not os.path.exists(output_path):
    os.mkdir(output_path)
print("Starting ")

with open(annotations_path, 'r') as csvfile:
    sign_annotations = csv.reader(csvfile, delimiter=',', quotechar='|')
    i=0
    
    for sign_info in sign_annotations:
        if i==0:
            header=sign_info
            i=1
            continue
        i=i+1
        new_gloss=sign_info[1]
        
        if not new_gloss=="" or sign_info[2]=="------------":
            continue
            
        session=sign_info[12]    
        scene=sign_info[13]
        
        video_filename="%s_%s_camera1.mov" % (session,scene)
        video_filepath=os.path.join(videos_path,video_filename)
        start=float(sign_info[14])
        end=float(sign_info[15])
        
        interpreter=sign_info[2]
        gloss=sign_info[3]
        data = skvideo.io.ffprobe(video_filepath)['video']
        video_duration=float(data['@duration'])
        rate_fraction = data['@r_frame_rate']
        num,den=rate_fraction.split("/")
        rate=float(num)/float(den)
        
        start_seconds=start/rate
        end_seconds=end/rate
#         print("%s rate %f from %f s to %f s" % (video_filepath,rate,start_seconds,end_seconds))
        
        
        output_filename="%06d_%s_%s_%s_%s_camera1.avi" % (i,gloss,interpreter,session,scene)
        output_filename=output_filename.replace("/","_")
        output_filepath=os.path.join(output_path,output_filename)
        
        duration=end_seconds-start_seconds
        if duration<min_duration:
            missing=min_duration-duration
            extension=missing/2
            start_seconds=max(start_seconds-extension,0)
            end_seconds=min(end_seconds+extension,video_duration)
            duration=end_seconds-start_seconds
        command=["ffmpeg", "-ss %f" % start_seconds, "-i %s" % video_filepath, "-c:v libx264 -c:a aac", "-t %f" % duration, output_filepath]
        command_string=" ".join(command)
        print(command_string)
#         subprocess.call(command)
        os.system(command_string)
        

        
```
